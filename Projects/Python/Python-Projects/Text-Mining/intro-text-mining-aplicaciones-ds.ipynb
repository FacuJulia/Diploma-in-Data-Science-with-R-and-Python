# 2. Preprocesamiento

Transformar el/los texto/s con los que se este trabajando (corpus), que se encuentran en lenguaje humano a un formato legible para las máquinas.

## # Tokenizar: separar texto en palabras

Consiste en dividir el texto en entidades significativas (palabras, oraciones, caracteres, N-gramas).

```python
from nltk import RegexpTokenizer
pattern1 = r'\w+'
tokenizer = RegexpTokenizer(pattern1)

tokens_texto = tokenizer.tokenize(texto) 
print(tokens_texto)
```

### Diccionario de frecuencias - Conteo de palabras

```python
from nltk import FreqDist
import pandas as pd

frecuencia = FreqDist(tokens_texto) 
frecuencia

#df = pd.DataFrame([[key, frecuencia[key]] for key in frecuencia.keys()], columns=['Token', 'Frecuencia'])
#print(df.sort_values(by='Frecuencia', ascending=False)) # Imprimo ordenando por frecuencia (invertido)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
frecuencia.plot(20, cumulative=False)
```

## # Normalización

Consiste en eliminar mayúsculas, minúsculas, puntación, números, etc.

```python
for t in tokens_texto:
    t = t.lower()
```

## # Eliminar STOP WORDS (No aportan significado)

Consiste en eliminar el ruido, palabras comunes a todos los textos, para reducir la dimensionalidad de los datos y simplificar el modelo. 

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('spanish'))
#stop_words

#remover = {'algunas', 'a', 'en'} #Escribir aquí las palabras que se quiere sacar de la lista de stopwords de nltk
#stop_words_modificadas = set([word for word in stop_words if word not in remover])
#stop_words = ['de','la','para','del','El','que','las','se','por','uno','en','lo','un', 'los']

tokens_filtrados = []
 
for t in tokens_texto:
    t = t.lower()
    if t not in stop_words:
        tokens_filtrados.append(t)

frec_filtrada = FreqDist(tokens_filtrados)      
frec_filtrada  

#df = pd.DataFrame([[key, frec_filtrada[key]] for key in frec_filtrada.keys()], columns=['Token', 'Frecuencia'])
#print(df.sort_values(by='Frecuencia', ascending=False)) # Imprimo ordenando por frecuencia (invertido)
```

```python
set(tokens_filtrados)

plt.figure(figsize=(14, 3))

fdist2 = FreqDist(tokens_filtrados) 
fdist2.plot(50)
```

## # Stemming

Consiste en distinguir tokens semánticamente relevantes. Para la libreria NLTK es el módulo **Snowball** que permite reducir las palabras a la raiz.
 - **Chang**ing
 - **Chang**ed
 - **Chang**e

```python
import re
pattern = re.compile(r'proce.*')

proces = []
for token in tokens_texto:
    proces.extend(re.findall(pattern, token))

print(proces[0:20])
```

## # Vectorización

Es el ultimo paso del preprocesamiento, consiste en el mapeo de tokens a un vector de números reales, que se utiliza como entrada de un modelo de ML. Para eso se genera una matriz dispersa donde cada fila es un documento y cada columna una palabra.

El método mas simple es contar las palabras (Bag of Words). Otro método es otorgar un valor a cada término de acuerdo a su frecuencia en el documento sobre la cantidad total de documentos que aparece la palabra (TF- IDF)

**Bag of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
 
vec=CountVectorizer()
#vec.get_params()

corpus = [
    'This is the first first document.',
    'This is the second second document.',
    'And And And the third one.',
    'Is this the first document?',
]

X = vec.fit_transform(corpus)

print("Features: ", vec.get_feature_names())
 
X.toarray() 
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

text = [
    "The quick brown fox jumped over the lazy dog.",
    "The dog.",
    "The fox"]

vectorizer.fit(text)

print(vectorizer.vocabulary_)
```

```python
vector = vectorizer.transform(text)

print(vector.toarray())
```
